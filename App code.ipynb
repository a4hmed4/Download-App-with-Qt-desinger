```python
from PyQt5.QtGui import *
from PyQt5.QtCore import *
from PyQt5.QtWidgets import *
import sys
from PyQt5.uic import loadUiType
import urllib.request

# Load UI
ui, _ = loadUiType('task 1 Interactive.ui')

class MainApp(QMainWindow, ui):
    def __init__(self, parent=None):
        super(MainApp, self).__init__(parent)
        self.setupUi(self)
        self.Handle_UI()
        self.Handle_Buttons()

    def Handle_UI(self):
        # Set window properties
        self.setWindowTitle('My Download App')
        self.setFixedSize(665, 440)

    def Handle_Buttons(self):
        # Connect buttons to their respective methods
        self.pushButton.clicked.connect(self.download)         # Download button
        self.pushButton_2.clicked.connect(self.Handle_Browse)  # Browse button

    def Handle_Browse(self):
        # Open dialog to select save location
        save_location, _ = QFileDialog.getSaveFileName(self, "Save As", "", "All Files (*.*)")
        
        # Set the chosen save location in the lineEdit_2 if selected
        if save_location:
            self.lineEdit_2.setText(save_location)

    def download(self):
        # Get download URL and save location from the input fields
        download_url = self.lineEdit.text()
        save_location = self.lineEdit_2.text()

        # Validate the inputs
        if not download_url:
            QMessageBox.warning(self, "Input Error", "Please provide a download URL.")
            return
        if not save_location:
            QMessageBox.warning(self, "Input Error", "Please select a save location.")
            return

        # Reset the progress bar before starting
        self.progressBar.setValue(0)

        try:
            # Download file and show progress
            urllib.request.urlretrieve(download_url, save_location, self.Handle_Progress)
        except urllib.error.URLError as e:
            QMessageBox.warning(self, "Download Error", f"Failed to download the file. Error: {str(e)}")
            return
        except Exception as e:
            QMessageBox.warning(self, "Download Error", f"An error occurred during the download: {str(e)}")
            return

        # Show download success message and reset the fields
        QMessageBox.information(self, "Download Completed", "The download completed successfully.")
        self.lineEdit.setText('')
        self.lineEdit_2.setText('')
        self.progressBar.setValue(0)

    def Handle_Progress(self, blocknum, blockSize, totalSize):
        # Calculate progress and update progress bar
        read = blocknum * blockSize
        if totalSize > 0:
            percent = read * 100 / totalSize
            self.progressBar.setValue(int(percent))
            QApplication.processEvents()  # Ensure the UI updates during download


def main():
    app = QApplication(sys.argv)
    window = MainApp()
    window.show()
    app.exec_()

if __name__ == '__main__':
    main()
```

```python
from PyQt5.QtGui import *
from PyQt5.QtCore import *
from PyQt5.QtWidgets import *
import sys
from PyQt5.uic import loadUiType
import urllib.request

ui, _ = loadUiType('task 1 Interactive.ui')


class MainApp(QMainWindow, ui):
    def __init__(self, parent=None):
        super(MainApp, self).__init__(parent)
        self.setupUi(self)
        self.Handle_UI()
        self.Handle_Buttons()

    def Handle_UI(self):
        self.setWindowTitle('My download app')
        self.setFixedSize(565, 440)

    def Handle_Buttons(self):
        self.pushButton.clicked.connect(self.download)       
        self.pushButton_2.clicked.connect(self.Handle_Browse)    

    def Handle_Browse(self):

        save_location, _ = QFileDialog.getSaveFileName(self, "Save As", "", "All Files (*.*)")
        
        if save_location:
            self.lineEdit_2.setText(save_location)

    def download(self):
        print('Start Download')
        download_url = self.lineEdit.text()
        save_location = self.lineEdit_2.text()

        try:
            urllib.request.urlretrieve(download_url, save_location, self.Handle_Progress)
        except urllib.error.URLError as e:
            QMessageBox.warning(self, "Download Error", f"Failed to download the file. Error: {str(e)}")
            return
        except Exception as e:
            QMessageBox.warning(self, "Download Error", f"An error occurred during the download: {str(e)}")
            return

        QMessageBox.information(self, "Download Completed", "The download completed successfully.")
        self.lineEdit.setText('')
        self.lineEdit_2.setText('')
        self.progressBar.setValue(0)

    def Handle_Progress(self, blocknum, blockSize, totalSize):
        read = blocknum * blockSize
        if totalSize > 0:
            percent = read * 100 / totalSize
            self.progressBar.setValue(int(percent))
            QApplication.processEvents()


def main():
    app = QApplication(sys.argv)
    window = MainApp()
    window.show()
    app.exec_()


if __name__ == '__main__':
    main()
```
